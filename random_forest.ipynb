```python
import pandas as pd
from sklearn.datasets import load_digits, load_iris

dig = load_digits()
dir(dig)
```

```python
import matplotlib.pyplot as plt

%matplotlib inline
plt.gray()
for i in range(4):
  plt.matshow(dig.images[i])
```

```python
df = pd.DataFrame(dig.data, columns = dig.feature_names )
df['digit'] = dig.target
```

```python
from sklearn.model_selection import train_test_split

x = df.drop(columns = 'digit')
y = df.digit

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(x_train, y_train)
```

```python
y_pred = rfc.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
import seaborn as sns

plt.figure(figsize = (10,7))
sns.heatmap(cm, annot = True, cmap = 'Reds')
plt.xlabel('predicted')
plt.ylabel('actual')
plt.show()
```

```python
iris = load_iris()

df = pd.DataFrame(iris.data, columns = iris.feature_names)
df['type'] = iris.target
df['type_name'] = df['type'].apply(lambda x : iris.target_names[x])
df
```

```python
x = df.drop(columns = ['type', 'type_name'])
y = df['type']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)
```

```python
model = RandomForestClassifier()

model.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score as acc

y_pred = model.predict(x_test)

acc(y_test, y_pred)
con = confusion_matrix(y_test, y_pred)
```

```python
plt.figure(figsize = (10,7))
sns.heatmap(con, annot = True, cmap = 'Blues')
plt.xlabel('predicted iris')
plt.ylabel('actual iris')
plt.show()
```
