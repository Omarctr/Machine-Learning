```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
dir(iris)
```

```python
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df['iris'] = iris.target
df['name'] = df.iris.apply(lambda x : iris.target_names[x])
df
```

```python
from sklearn.model_selection import train_test_split
x = df.drop(columns = ['iris', 'name'])
y = df.iris

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.svm import SVC
svc = SVC(kernel = 'rbf', C = 30, gamma = 'auto')
svc.fit(x_train, y_train)
svc.score(x_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score

cross_val_score(SVC(kernel = 'linear', C = 30), x, y, cv = 5)
```

```python
cross_val_score(SVC(kernel = 'poly', C = 10, gamma = 'auto'), x, y, cv = 5)
```

```python
cross_val_score(SVC(kernel = 'rbf', C = 10), x, y, cv = 5)
```

# **other way**

```python
from sklearn.model_selection import GridSearchCV

clf = GridSearchCV(SVC(gamma = 'auto'), {
    'C' : [1, 10, 20, 30],
    'kernel' : ['rbf', 'linear', 'poly']
}, cv = 5)
```

```python
clf.fit(x, y)
clf.cv_results_
```

```python
res = pd.DataFrame(clf.cv_results_)
res
```

```python
clf.best_params_
```

```python
from sklearn.model_selection import RandomizedSearchCV

rs = RandomizedSearchCV(SVC(gamma = 'auto'), {
    'C' : [1, 10, 20],
    'kernel' : ['rbf', 'poly', 'linear']
}, cv = 5, n_iter = 2)

rs.fit(x, y)
result = pd.DataFrame(rs.cv_results_)
result
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier


model_params = {
    'svc' : {'model' : SVC(),
            'params' : {'C' : [10, 20, 30],
                        'kernel' : ['rbf', 'linear']}
             },
    'random forest' : {'model' : RandomForestClassifier(),
                       'params' : {'n_estimators' : [100, 150, 200]}
                       },
    'logistic regression' : {'model' : LogisticRegression(max_iter = 1000),
                             'params' : {'C' : [10, 20, 30]}
                             }
}

scores = []
for model_name, vals in model_params.items():
  mod = GridSearchCV(vals['model'], vals['params'], cv = 5, return_train_score = False)
  mod.fit(x, y)
  scores.append({
      'model' : model_name,
      'best_score' : mod.best_score_,
      'best_parameters' : mod.best_params_
  })


model_table = pd.DataFrame(scores)
model_table
```

```python
from sklearn.datasets import load_digits

dig = load_digits()
df = pd.DataFrame(dig.data, columns = dig.feature_names)
df['digit'] = dig.target
```

```python
df
x = df.drop(columns = 'digit')
y = df.digit
```

```python
from sklearn.naive_bayes import GaussianNB, BernoulliNB, MultinomialNB
model_params = {
    'svc' : {'model' : SVC(),
            'params' : {'C' : [10, 20, 30],
                        'kernel' : ['rbf', 'linear']}
             },
    'random forest' : {'model' : RandomForestClassifier(),
                       'params' : {'n_estimators' : [100, 150, 200]}
                       },
    'logistic regression' : {'model' : LogisticRegression(max_iter = 1000),
                             'params' : {'C' : [10, 20, 30]}
                             },
    'gaussian naive bayes' : {'model' : GaussianNB(),
                              'params' : {'var_smoothing' : [1e-9]}},
    'bernoulli naive bayes' : {'model' : BernoulliNB(),
                               'params' :{'fit_prior' : [False, True],
                                          'force_alpha' : [False, True]}},
    'multinomial naive bayes' : {'model' : MultinomialNB(),
                                 'params' : {'fit_prior' : [False, True],
                                             'alpha' : [1, 5, 10 ,20]}}
}

adv_scores = []
for model_name , val in model_params.items():
  clf = GridSearchCV(val['model'], val['params'], cv = 5, return_train_score = False)
  clf.fit(x, y)
  adv_scores.append({
      'model_name' : model_name,
      'best_score' : clf.best_score_,
      'best_parameters' : clf.best_params_
  })


  model_table = pd.DataFrame(adv_scores)
  model_table
```

```python
model_table
```
