```python
import pandas as pd
df = pd.read_csv("/content/titanic.csv")
df.head()
```

```python
from numpy import True_
df.drop(columns = ['Name','SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], inplace = True)
df
```

```python
temp = pd.get_dummies(df.Sex)
df = pd.concat([df, temp], axis = 1)
df
```

```python
df.drop(columns = 'Sex', inplace = True)
df
```

```python
df.Age = df.Age.fillna(df.Age.mean())
df
```

```python
from sklearn.model_selection import train_test_split

x = df.drop(columns = 'Survived')
y = df.Survived

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
y_test[:10]
```

```python
model.predict(x_test[:10])
```

```python
model.predict_proba(x_test[:10])
```

```python
em = pd.read_csv("/content/spam.csv")
em.head()
```

```python
em.groupby('Category').describe()
```

```python
em['spam'] = em.Category.map({'spam' : 1, 'ham' : 0})
em
```

```python
x = em.Message
y = em.spam

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()
x_train_c = cv.fit_transform(x_train.values)
x_test_c = cv.transform(x_test.values)
```

```python
from sklearn.naive_bayes import MultinomialNB

mnb = MultinomialNB()
mnb.fit(x_train_c, y_train)
```

```python
emails = [
    'Hey mohan, can we get together to watch football game tomorrow?',
    'Upto 20% discount on parking, exculsive offer just for you. Dont miss this reward!'
]
label = [0,1]

text_to_num = cv.transform(emails)
mnb.score(text_to_num, label)
```

```python
mnb.score(x_test_c, y_test)
```

# **better way to transform text to numerical**

```python
from sklearn.pipeline import Pipeline

clf = Pipeline([
    ('vectorizer' , CountVectorizer()),
    ('nb', MultinomialNB())
])

clf.fit(x_train, y_train)
```

```python
clf.score(x_test, y_test)
```

```python
from sklearn.datasets import load_wine
w = load_wine()
wine = pd.DataFrame(w.data, columns = w.feature_names)
wine['class'] = w.target
wine
```

```python
wine.groupby('class').median()
```

```python
wine.drop(columns = ['ash', 'nonflavanoid_phenols', 'proanthocyanins',  'hue'], inplace = True)
```

```python
wine
```

```python
x = wine.drop(columns = 'class')
y = wine['class']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)
```

```python
gauss = Pipeline([
    ('nb', GaussianNB())
])

multi = Pipeline([
    ('nb', MultinomialNB())
])
gauss.fit(x_train, y_train)
multi.fit(x_train, y_train)
```

```python
print(gauss.score(x_test, y_test))
multi.score(x_test, y_test)
```
