```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.DataFrame({'area' : [2600, 3000, 3200, 3600, 4000],
                   'price' : [550000, 565000, 610000, 680000, 725000]})
df
```

```python
%matplotlib inline
plt.scatter(df.area, df.price)
plt.title('data frame')
plt.xlabel('area')
plt.ylabel('prices')
plt.show()
```

```python
rg = LinearRegression()
rg.fit(df[['area']], df.price)
```

```python
x = np.array(list(df['area']))
y = np.array(list(df['price']))
```

```python
def gradient(x, y):
  m_curr = b_curr = 0
  itr = 100
  n = len(x)
  lr = 0.001
  idx = 0
  for i in range(itr):
    y_pred = m_curr * x + b_curr
    plt.plot(x, y_pred)
    error = y - y_pred
    errorSq = (error)**2
    m_der = -(2/n) * sum(x * error)
    b_der = -(2/n) * sum(error)
    m_curr -= lr * m_der
    b_curr -= lr * b_der
    MSE = (1/n) * sum(errorSq)
    print(f'm: {m_curr}, b: {b_curr}, cost: {MSE}')

gradient(x, y)
```

```python
# Y = m (gradient) * x + b (intercept)
print(135.78767123*5000+180616.43835616432)
rg.predict([[5000]])
```
