```python
import pandas as pd
df = pd.read_csv('/homeprices (1).csv')
df
```

```python
town = pd.get_dummies(df.town)
```

```python
merged = pd.concat([df, town], axis = 1)
merged = merged.drop(['town', 'west windsor'], axis = 1)
```

```python
df
```

```python
from sklearn.linear_model import LinearRegression as lr
reg = lr()
x = merged.drop(columns = ['price'])
y = merged['price']
reg.fit(x, y)
```

```python
reg.predict([[2800, 0, 1]])
```

```python
reg.predict([[3400, 0,0 ]])
```

```python
print(f'{reg.score(x, y) *100:.2f}%')
```
